```python
import json
import os
from collections import Counter
PROJECT_BASE_PATH = os.path.join(
            os.path.expanduser("~"),"Desktop", "email_search_project")

EMAIL_IMAGES_DIR = os.path.join(PROJECT_BASE_PATH, "example_emails")
EMAIL_METADATA_DIR = os.path.join(PROJECT_BASE_PATH, "email_metadata")

```

```python
layout_terms = Counter()
brand_style_terms = Counter()
text_terms = Counter()
color_scheme_terms = Counter()
other_terms = Counter()

for file in os.listdir(EMAIL_METADATA_DIR):

    if file.startswith("."):
        continue
    
    with open(os.path.join(EMAIL_METADATA_DIR, file), "r") as f:
        email_metadata_json = json.load(f)

    layout_terms.update(email_metadata_json['layout'])
    brand_style_terms.update(email_metadata_json['brand_style']) 
 
    layout_terms.update(email_metadata_json['layout'])
    brand_style_terms.update(email_metadata_json['brand_style'])
    text_terms.update(email_metadata_json['text'])
    color_scheme_terms.update(email_metadata_json['color_scheme'])
    other_terms.update(email_metadata_json['other'])

# Print the most common terms for each category
print("\nMost common layout terms:")
for term, count in layout_terms.most_common(40):
    print(f"{term}: {count}")

print("\nMost common brand style terms:")
for term, count in brand_style_terms.most_common(40):
    print(f"{term}: {count}")

print("\nMost common text terms:")
for term, count in text_terms.most_common(40):
    print(f"{term}: {count}")

print("\nMost common color scheme terms:")
for term, count in color_scheme_terms.most_common(40):
    print(f"{term}: {count}")

print("\nMost common other terms:")
for term, count in other_terms.most_common(40):
    print(f"{term}: {count}")
    
```
